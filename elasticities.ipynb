```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Plots comparing linear and constant demand elasticties, to show in the report

```python
# Choose nominal operating point (price and quantity)
l_hat = 150 # $
P_hat = 1 # MW
```

```python
P = np.linspace(0.1,1.5,100) # Define quantity vector
e = [-0.7,-0.3] # Choose some elasticities 
fig, ax = plt.subplots()
c = sns.color_palette()
i = 0
for ei in e:
    
    # Linear model
    du = l_hat + (l_hat/ei*P_hat)*(P-P_hat)
    ax.plot(P, du, '--', color=c[i])
    
    # Constant model
    K = l_hat/(P_hat**(1/ei)) # Choose K s.t. du(P_hat) = l_hat
    du = K*(P**(1/ei))
    ax.plot(P, du, label=ei, color=c[i])
    i += 1

ax.plot(P_hat, l_hat, 'r*')
ax.legend()
ax.get_legend().set_title('Elasticity')
ax.set_xlabel('MW')
ax.set_ylabel('Price')
ax.set_ylim([0, max(du)])
ax.set_ylim([0,1000])
#ax.set_title('Comparison of linear and constant demand elasticities')
plt.show()
```
